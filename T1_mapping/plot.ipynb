```python
import matplotlib.lines as mlines
fig= plt.figure(figsize=(7,6), dpi = 200, )
ax = plt.gca()
colors = ['lightcoral','orange','limegreen','royalblue']
markers = ['.','o','s','^']
labels = ['Measured','Calculated', "Bayes t", "Bayes Gauss"]
T1 = [800, 1200, 1600]
SNR = [20, 30, 40]
plt.xlim()
plt.ylim(10,70)
for k in T1:
    for j_n, j in enumerate([20,30,40]):
        T1_estimate, T1_sd_estimate, T1_sd_raw_estimate, T1_estimate_bayes, T1_sd_estimate_bayes, T1_estimate_bayes_Gauss, T1_sd_estimate_bayes_Gauss= np.load(f'metric_{k}_2_{j}_2048.npy', allow_pickle = True).item().values()
        y = [T1_estimate.std(),T1_sd_estimate.mean(),T1_sd_estimate_bayes.mean(),T1_sd_estimate_bayes_Gauss.mean()]
        x = [k]*len(y)
        for i in range(len(x)):
            if i== 0:
                plt.scatter(x[i],y[i],s = 50,marker = markers[i],color = colors[j_n], label = labels[i], alpha = 1)
            else: 
                plt.scatter(x[i],y[i],s = 50,marker = markers[i],color = colors[j_n], label = labels[i], alpha = 1, facecolor="None")

legend_lines_2 = []
for i in range(len(T1)):
    legend_lines_2.append(mlines.Line2D([], [], color=colors[i], 
                             label=SNR[i],lw = 1))           


legend_1= plt.legend(handles = legend_lines_2, loc = (0.013,0.6))


legend_lines = []
for i in range(len(labels)):
    legend_lines.append(mlines.Line2D([], [], color='b', marker=markers[i],
                            markersize=6, label=labels[i],lw = 0))                      
ax.legend(handles=legend_lines, loc = (0.013, 0.78))


plt.gca().add_artist(legend_1)

plt.xlabel('T1 (ms)')
plt.ylabel('T1 SD estimates (ms)')
```
